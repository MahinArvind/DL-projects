fft -> spectrum

```python
#power spectrum1
fft1 = np.fft.fft(signal1)
magnitude1 = np.abs(fft1)
frequency1 = np.linspace(0,sr1, len(magnitude1))


plt.plot(frequency1, magnitude1)
plt.title('Covid Mild')
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.show()

#power spectrum2

fft2 = np.fft.fft(signal2)
magnitude2 = np.abs(fft2)
frequency2 = np.linspace(0,sr2, len(magnitude2))

plt.plot(frequency2, magnitude2)
plt.title('Healthy')
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.show()
```

STFT-> spectogram

```python
n_fft = 2048 #no. samples
hop_length =512

stft1 = librosa.core.stft(signal1, hop_length= hop_length, n_fft= n_fft)
spectrogram1 = np.abs(stft1)

log_spectrogram1 = librosa.amplitude_to_db(spectrogram1)
#replace log_spectrogram with spectrogram for normal spectogram
librosa.display.specshow(log_spectrogram1, sr = sr1, hop_length=hop_length)
plt.title("Covid Mild")
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.show()


stft2 = librosa.core.stft(signal2, hop_length= hop_length, n_fft= n_fft)
spectrogram2 = np.abs(stft2)

log_spectrogram2 = librosa.amplitude_to_db(spectrogram2)

librosa.display.specshow(log_spectrogram2, sr = sr2, hop_length=hop_length)
plt.title("Healthy")
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.show()
```
